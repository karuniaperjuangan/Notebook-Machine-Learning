```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import tree
np.set_printoptions(precision=3, suppress=False)
```

```python
iris = datasets.load_iris( as_frame=True)
```

```python
iris.data
```

```python
x_train, x_test, y_train, y_test = train_test_split(iris.data,iris.target, random_state = 42)
```

```python
clf = tree.DecisionTreeClassifier(max_depth =3, random_state = 0)
clf.fit(x_train, y_train)
```

```python
plt.figure(figsize=(20,20), facecolor ='w')
tree.plot_tree(clf, feature_names = iris.feature_names, class_names = iris.target_names, filled = True, rounded=True, fontsize=16)
plt.show()
```

```python
predict_test = clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cf = confusion_matrix(y_test, predict_test, normalize='true')
sns.heatmap(cf, annot=True)
print(cf)
```
